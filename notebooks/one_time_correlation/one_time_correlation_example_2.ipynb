## Get the average value of the images and plot it

```python
def avg_img(img_array, num_images):
    """
    This will provide the average value of the images
    Parameters
    ----------
    img_array: ndarray
        array of stack of images for one time correlation

    n_start: int
        starting image number

    num_images: int
        number of images

    Returns
    ------
    avg_img : array

    """
    #n_end = n_start + num_images
    avg_img = 0
    for n in range(num_images):
        avg_img += img_array[n]

    avg_img /=float(num_images)
    return avg_img
```

```python
avg_img = avg_img(img_stack, img_stack.shape[0])  
```

```python
fig, ax = plt.subplots()
plt.imshow(avg_img)
plt.show()
```

## Plot the roi's 

```python
def test_demo(ax, inds, pix_list, img_dim, image):
    """
    This will plot the reqiured roi's on the image
    """
    tt = np.zeros(img_dim).ravel() * np.nan
    tt[pix_list] = inds

    vmin, vmax = np.percentile(image, [100, 100])
    im = ax.imshow(image, interpolation='none', norm=LogNorm(), vmin=vmin, vmax=vmax)
    im = ax.imshow(tt.reshape(*img_dim), cmap='Paired', interpolation='nearest')
```

```python
tt = np.zeros(img_stack[0].shape).ravel()
tt[pix] = qind
```

```python
fig, ax = plt.subplots()
plt.title("NIPA_GEL_50K")
test_demo(ax, qind, pix, img_stack[0].shape, avg_img)
plt.show()
```

## Use number of levels 7 and number of buffers or channles 8 for multi-tau correlation

```python
num_qs = 11

num_levels = 7
num_bufs = 8
```

```python
g2, lag_steps, elapsed_time = corr.auto_corr(num_levels, num_bufs, num_qs,
                                             num_pixels, pix, qind, img_stack)
```

```python
elapsed_time
```

```python
lag_time = 0.001
lag_step = lag_steps[:g2.shape[0]]
lags = lag_step*lag_time
```

## Plot the g2's

```python
if num_qs%2 == 0:
    y = num_qs/2
else:
    y = (num_qs/2 +1)
fig = plt.figure()
    #plt.title("g2")
for i in range(num_qs):
    ax = fig.add_subplot(2, y, i+1)
    ax.set_ylabel("g2")
    ax.set_xlabel("lags")
    ax.set_xscale('log')
    fig.subplots_adjust(left=0.2, wspace=1, hspace=0.2)
    plt.semilogx(lags, g2[:, i], '-o')
plt.show()
```

```python
plt.semilogx(lags, g2[:,0], '-o')
plt.title("NIPA_Gels")
plt.xlabel("lags")
plt.ylabel("g2")
plt.show()
```
