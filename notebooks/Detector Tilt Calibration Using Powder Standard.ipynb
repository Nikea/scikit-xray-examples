# Detector Tilt Calibration Using Powder Standard

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pims
import skimage.filter
import skxray.calibration
import skxray.plotting
```

```python
img = pims.ImageSequence(['../data/powder-diffraction-images/LaB6-00001.tif'], dtype=np.uint16)[0]

sample_distance = 208  # mm
pixel_size = (.200, .200)  # um
wavelength = 0.1839  # Angstroms or nm -- how does our code expect it?!
```

```python
img
```

```python
blurred_img = skimage.filter.gaussian_filter(img, 1)
center = skxray.calibration.find_ring_center(blurred_img)
```

```python
center
```

```python
r = skxray.core.pixel_to_radius(img.shape, center, pixel_size)
edges, val, count = skxray.calibration.bin_1D(r.ravel(), img.ravel().astype('float'), nx=2000)
```

```python
fig, ax = plt.subplots()
skxray.plotting.lognorm_imshow(img, ax=ax)
skxray.plotting.draw_rings([], center, ax=ax)
```

```python
two_theta = skxray.core.radius_to_twotheta(sample_distance, edges)
q = skxray.core.twotheta_to_q(two_theta, wavelength)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.step(edges[:-1], val/count)
ax2.step(edges[:-1], val/count)
ax2.set(yscale='log')
for ax in [ax1, ax2]:
    ax.set(xlabel='q [inverse Angstroms]', ylabel='intensity')
```
